## Hoja de Trabajo #1 - Jupyter Notebooks, Markdown y Git


Las herramientas de control de versiones son utiles para mantener un seguimiento ordenado de los cambios que se realizan en el código fuente. Además, una de las mayores ventajas de utilziar estas herramientas es que nos permite realizarun trabajo colaborativo. 

### ¿Por qué es útil un control de versiones?
Utilizar un control de versiones nos permite poder acceder a código que hemos realizado anteriormente sobre el mismo proyecto, si es necesario retroceder debido a algún error o cambio que hayamos codificado. En el trabajo colaborativo nos permite acceder a los cambios que realizan otras personas. 

### Git
Git es un software de control de versiones, desollador con el objetivo de incrementar la eficiencia y confiabilidad de las versiones de las aplicaciones y códigos. Es una herramienta muy flexible que permite trabajar con varias tecnologías y plataformas, siendo la mas popular GitHub.

### GitHub
Es una plataforma para alojar respositorios de proyectos utilizando el CVS Git. Permite a los desarrolladores  el trabajo simultaneo y colaborativo, realizar tantos cambios como sea posible y retroceder en la linea de tiempo de los cambios que se han hecho. 

#### Repositorio
Los repositorios son espacios virtuales en donde se almacenan los directorios y ramas en las cuales se divide el proyecto.

3. Agregar todos los archivos al repositorio
    - Add  
        Comando:  
        > `$git add .`  
          

    - Commit  
        Comando:  
        > `$ git commit -m "Primer commit del respositorio"`  
          

    - Push  
        Comando:  
        > `$ git push`  
          

    - Git pedira autenticación para subir los cambios al respositorio
